```python
import pandas as pd
```

```python
dates = pd.date_range("2022", "2032", freq="Y").year
dates
```

```python
df = pd.DataFrame(columns=["Revenue", "EBITDA", "Taxes",
                           "Cap_Ex", "Work_Cap",
                           "Cash_Flow", "Discount_Factor",
                           "Present_Value_fcf"],
                  index=dates)
df.loc[2022, "Revenue"] = 222 * (1.18)
df
```

```python
for i in range(1, len(df)):
    df.iloc[i]["Revenue"] = df.iloc[i - 1]["Revenue"] * (1.18)
```

```python
df
```

```python
df["EBITDA"] = df["Revenue"] * 0.45
```

```python
df["Taxes"] = -df["EBITDA"] * 0.21
```

```python
df["Cap_Ex"] = -df["Revenue"] * 0.03
```

```python
df["Work_Cap"] = -df["Revenue"] * 0.06
```

```python
df["Cash_Flow"] = df["EBITDA"] + df["Taxes"] + df["Cap_Ex"] + df["Work_Cap"]
```

```python
df
```

```python
df.iloc[0]["Discount_Factor"] = 0.94
```

```python
        for i in range(1, len(df)):
    df.iloc[i]["Discount_Factor"] = 1 / ((1 + 0.06) ** (i + 1))
df
```

```python
df["Present_Value_fcf"] = df["Cash_Flow"] * df["Discount_Factor"]
```

```python
df
```

```python
revenue_2032 = df.loc[2031, "Revenue"] * 1.18
ebitda_2032 = revenue_2032 * 0.45
terminal_value = ebitda_2032 * 32.70
present_value_of_project_fcf = df["Cash_Flow"].sum()
```
